### The gioi di dong

```python
import httpx
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.thegioididong.com/dtdd-apple-iphone'

#tạo một SSLContext

ssl_context = httpx.create_ssl_context()
ssl_context.options |= 0x4 # cho phép legacy renegotiation

# Gửi response với SSLContext

with httpx.Client(verify=ssl_context) as client:
    response = client.get(url, headers={'User-Agent': 'Mozilla/5.0'})

if response.status_code == 200:
    content = response.text
else:
    print(f'Lỗi: {response.status_code}')
    exit()

soup = BeautifulSoup(content,'html.parser')

print('done')
```

```python
# Tạo danh sách để lưu trữ thông tin sản phẩm
products = []

# Tìm tất cả các sản phẩm trong trang web
product_items = soup.find_all('li', class_='item')

for item in product_items:
    # Lấy thông tin sản phẩm
    product_id = item.get('data-id')
    product_name = item.find('h3').get('title') if item.find('h3') else None
    merged_name = item.find('div', class_='prods-group').get('data-mergename') if item.find('div', class_='prods-group') else None
    price = item.find('strong', class_='price').text if item.find('strong', class_='price') else None
    old_price = item.find('p', class_='price-old').text if item.find('p', class_='price-old') else None

    # Lấy mô tả sản phẩm
    utility = item.find('div', class_='utility')
    description = "\n".join([p.text for p in utility.find_all('p')]) if utility else None

    # Thêm thông tin sản phẩm vào danh sách
    products.append({
        'ID': product_id,
        'nam': product_name,
        'merged_name': merged_name,
        'price': price,
        'old_price': old_price,
        'desc': description
    })

# Tạo DataFrame từ danh sách sản phẩm
df = pd.DataFrame(products)

# Lưu DataFrame vào file CSV
df.to_csv('san_pham_dien_thoai.csv', index=False, encoding='utf-8-sig')

print("Dữ liệu đã được lưu vào file san_pham_dien_thoai.csv")
```

### Get data from Skin shop

```python
# url
url = 'https://www.happyskin.vn/duong-am'

#tạo một SSLContext

ssl_context = httpx.create_ssl_context()
ssl_context.options |= 0x4 # cho phép legacy renegotiation

# Gửi response với SSLContext

with httpx.Client(verify=ssl_context) as client:
    response = client.get(url, headers={'User-Agent': 'Mozilla/5.0'})

if response.status_code == 200:
    content = response.text
else:
    print(f'Lỗi: {response.status_code}')
    exit()

soup = BeautifulSoup(content,'html.parser')

print('done')
```

```python
# Tạo danh sách để lưu trữ thông tin sản phẩm
skin_products = []

# Tìm tất cả các sản phẩm trong trang web
product_items = soup.find_all('div', class_='product-square-item')

for item in product_items:
    # Lấy thông tin sản phẩm
    product_image_div = item.find('div', class_='product-image')
    product_id = product_image_div['data-id'] if product_image_div and 'data-id' in product_image_div.attrs else None
    product_name = item.find('div', {'data-name': True})['data-name'] if item.find('div', {'data-name': True}) else None
    price = item.find('span', class_='price').text.strip() if item.find('span', class_='price') else None
    old_price = item.find('span', class_='old-price').text.strip() if item.find('span', class_='old-price') else None

    # Lấy mô tả sản phẩm
    utility = item.find('div', class_='product-brief')
    description = "\n".join([p.text.strip() for p in utility.find_all('p')]) if utility else None

    # Thêm thông tin sản phẩm vào danh sách
    skin_products.append({
        'ID': product_id,
        'nam': product_name,
        'price': price,
        'old_price': old_price,
        'desc': description
    })

# Tạo DataFrame từ danh sách sản phẩm
df = pd.DataFrame(skin_products)

# Lưu DataFrame vào file CSV
df.to_csv('duong_am_da.csv', index=False, encoding='utf-8-sig')

print("Dữ liệu đã được lưu vào file duong_am_da.csv")
```
